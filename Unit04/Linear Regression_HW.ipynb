```python
#匯入在sklearn.linear_model套件裡面的LinearRegression模型
from sklearn.linear_model import LinearRegression

#匯入在sklearn.model_selection套件裡面的train_test_split模組
from sklearn.model_selection import train_test_split

#切割數據集(訓練資料80%、測試資料20%，設定random_state=20)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 20)

#畫出訓練資料集的matplotlib圖形

plt.scatter(x_train, y_train)
plt.show()
```

## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
# 1. 只用訓練資料集的資料進行linear regression演算法
LR = LinearRegression()
X_train = x_train.reshape(-1, 1)
LR.fit(X_train, y_train)
Y_train = LR.predict(X_train)

print("Intercept =", LR.intercept_)
print("Slope =", LR.coef_[0])
print("y_hat = {0}x + {1}".format(LR.coef_[0], LR.intercept_))
```

```python
# 2. 並計算出訓練階段的MSE
# from sklearn import metrics
# print(metrics.mean_squared_error(Y_train, y_train))

MSE = np.mean((y_train - Y_train)**2)
print("MSE =", MSE)
```

```python
# 3. 畫出目標點(藍色)與預測點(紅色)的對應圖 

plt.scatter(x_train, y_train)
plt.plot(x_train, Y_train, 'r')
plt.show()
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>
使用X = 2.44897959，預測出來數值應該為 4.3025375<br>
所有測試集資料
$$\widehat{y}=xw+b=w_{1}x_{1}+b$$

```python
print("y_hat = {0}x + {1}".format(LR.coef_[0], LR.intercept_))

# 測試資料集的資料預測
Y_test = LR.predict(x_test.reshape(-1, 1))

```

```python
# if X = 2.44897959, then predict values should be 4.3025375
X = [[2.44897959]]

print("if X = {0}, then predict values = {1}".format(X[0][0], LR.predict(X)[0]))
```

2. 並計算出測試階段的MSE

```python
MSE_test = np.mean((y_test - Y_test) ** 2)
print("MSE for test dataset =", MSE_test)

# metrics.mean_squared_error(y_test, Y_test)
```

3. 畫出目標點(藍色)與預測點(紅色)的對應圖

```python

plt.title("f(x) = {0:.1f}x + {1:.1f}, mse:{2:.2f}".format(LR.coef_[0], LR.intercept_, MSE_test))
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x_test, y_test, label = 'Ground Truth')
plt.scatter(x_test, Y_test, label = 'Predict')
plt.legend()
plt.show()
```
